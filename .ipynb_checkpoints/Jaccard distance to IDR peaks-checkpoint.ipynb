```python
def g4_df_to_bed(g4_df):
    g4_df['start-end']=g4_df['Genomic intervals of rG4'].str.split(':', expand = True)[1]
    g4_df['start']=g4_df['start-end'].str.split('-', expand = True)[0]
    g4_df['end']=g4_df['start-end'].str.split('-', expand = True)[1]
    
    return BedTool.from_dataframe(g4_df[['chromosome', 'start', 'end', 'Overlapping Gencode gene names', 'rG4 structural class','strand']])
```

```python

def get_contingency(g4,idr, region = utr):
    '''
    [idr & G4, IDR not G4],
    [G4 not IDR, neither]
    
    '''
    # select IDRs and G4 in UTR only
    idr_in_utr = idr.intersect(region, s = True, u = True)
    g4_in_utr = g4.intersect(region, s = True, u = True)
    
    
    intersect = len(idr_in_utr.intersect(g4_in_utr, s = True, u = True))
    not_rg4 = len(idr_in_utr)-intersect
    not_idr = len(g4_in_utr)-intersect
    neither = len(region) - not_rg4 - not_idr - intersect
    
    return intersect, not_rg4, not_idr, neither
```

```python
from scipy.stats import fisher_exact
from statsmodels.stats.multitest import fdrcorrection
def get_G4_enrichment(region = utr):
    ''' calculate fisher exact test statistics for all RBPs'''
    ans = []
    
    for index, row in master_df.iterrows():
        try:
            idr_path = row['idr']
            cell_line = row['Cell line']
            rbp = row['RBP']
            uid = row['uid']
            idr = BedTool(idr_path)
            intersect, not_rg4, not_idr, neither = get_contingency(rg4_regions, idr, region = region)
            contingency = np.array([intersect, not_rg4, not_idr, neither]).reshape(2,2)
            odds, pval = fisher_exact(contingency, alternative = 'greater')
            ans.append([uid, cell_line, intersect, not_rg4, not_idr, neither, odds, pval])
        except Exception as e:
            print(uid, e)

    idr_stats = pd.DataFrame(ans, columns = ['uid', 'cell line', 'n_intersect', 'idr_only', 'rg4_only', 'neither', 'odds', 'pval'])
    idr_stats=pd.merge(idr_stats, master_df[['uid', 'RBP']], left_on = 'uid', right_on = 'uid')
    idr_stats['-log10_pval'] = -np.log10(idr_stats['pval'])
    idr_stats['log2_odds'] = np.log2(idr_stats['odds'])
    _, idr_stats['FDR'] = fdrcorrection(idr_stats['pval'].tolist())
    
    return idr_stats
```

```python
def plot_test_result(idr_stats, fdr_thres = 0.1):
    idr_stats.plot.scatter(x = 'log2_odds', y = '-log10_pval', figsize = (10,10))

    for index, row in idr_stats.loc[idr_stats['FDR']<fdr_thres].iterrows():
        plt.text(y=row['-log10_pval'], x=row['log2_odds'], s=row['RBP']+'_'+row['cell line'])
```

```python
utr_stats = get_G4_enrichment(region = utr)
```

```python
plot_test_result(utr_stats)
plt.title('G4 enrichment in UTR')
```

```python
intron_stats = get_G4_enrichment(region = intron)
plot_test_result(intron_stats)
plt.title('G4 enrichment in intron')
```

# Preference in structural class

```python
potential_rg4_interacting = set(utr_stats.loc[(utr_stats['FDR']<0.1)&(utr_stats['odds']>2), 'uid']).union(
set(intron_stats.loc[(intron_stats['FDR']<0.1)&(intron_stats['odds']>2), 'uid']))
```

```python
from collections import Counter
type_dist = []
for uid in potential_rg4_interacting:
    idr_path = master_df.loc[master_df['uid']==uid, 'idr'].iloc[0]
    idr = BedTool(idr_path)
    
    # g4_intersect with idr
    r = rg4_regions.intersect(idr, s = True, u = True)
    
    # types
    types = Counter([i[4] for i in r])
    type_dist.append(types)
```

```python
type_dist_df = pd.DataFrame(type_dist, index = potential_rg4_interacting).fillna(0)
```

```python
type_dist_df.head()
```

```python
type_dist_df['RBP'] = type_dist_df.index.map(master_df.set_index('uid')['RBP'])
```

```python
struct_types = type_dist_df.columns[:-1]
```

```python
import seaborn as sns
d = type_dist_df[struct_types].divide(type_dist_df[struct_types].sum(axis = 1), axis = 0)
sns.clustermap(d, yticklabels = type_dist_df['RBP'], metric = 'correlation', z_score =1, cmap = 'RdBu_r', figsize = (5, 12))
```
